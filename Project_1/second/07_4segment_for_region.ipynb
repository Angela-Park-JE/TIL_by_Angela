```python
n_states = ['AM', 'AP', 'AC', 'TO', 'PA', 'RR', 'RO']
ne_states = ['MA', 'BA', 'SE', 'CE', 'AL', 'PB', 'PE', 'PI', 'RN']
mw_states = ['DF', 'GO', 'MT', 'MS']
se_states = ['RJ', 'MG', 'SP', 'ES']
s_states = ['SC', 'PR', 'RS']
```

```python
def region(x):
    if x in n_states:
        return 'n_states'
    elif x in ne_states:
        return 'ne_states'
    elif x in mw_states:
        return 'mw_states'
    elif x in se_states:
        return 'se_states'
    elif x in s_states:
        return 's_states'
    else:
        return 'none'
```

```python
# df['실험1'] = df['seller_state'].apply(lambda x : region(x))
# df['실험1'].nunique()
```

```python
df['seller_region'] = df['seller_state'].apply(lambda x : region(x))
df['customer_region'] = df['customer_state'].apply(lambda x : region(x))
df.sample(5)
```

```python
if df['seller_city']==df['customer_city']:
    df['region_div1'] = 'sc'
elif df['seller_city']!=df['customer_city'] and df['seller_state']==df['customer_state']:
    df['region_div1'] = 'ss'
elif df['seller_state']!=df['customer_state'] and df['seller_region']==df['customer_region']:
    df['region_div1'] = 'sr'
elif df['seller_region']!=df['customer_region']:
    df['region_div1'] = 'dr'
else:
    df['region_div1'] = 'none'
```

```python
df["reg_div"] = df[(df['customer_city'] == df['seller_city'])&(df['customer_state'] == df['seller_state'])]
```

```python
df.loc[(df['customer_city'] == df['seller_city'])&(df['customer_state'] == df['seller_state']) , 'region_div1'] = 'sc'
df.loc[(df['seller_city']!=df['customer_city'])&(df['seller_state']==df['customer_state']), 'region_div1'] = 'ss'
df.loc[(df['seller_state']!=df['customer_state'])&(df['seller_region']==df['customer_region']), 'region_div1'] = 'sr'
df.loc[(df['seller_region']!=df['customer_region']), 'region_div1'] = 'dr'
df['region_div1'].fillna(value='none', inplace=True)

df['region_div1'].unique()
```
